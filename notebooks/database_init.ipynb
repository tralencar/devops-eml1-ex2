```python
import firebase_admin
from firebase_admin import credentials, firestore
import pandas as pd
import os
```

```python
# Caminho relativo do notebook para a pasta .env
CREDENTIAL_PATH = os.path.join('..', '.env', 'serviceAccountKey.json')
```

```python
# Inicializa o Firebase Admin com a chave da pasta .env
if not firebase_admin._apps:
    cred = credentials.Certificate(str(CREDENTIAL_PATH))
    firebase_admin.initialize_app(cred)
```

```python
# Inicializa o Firestore
db = firestore.client()
```

```python
# Caminho relativo do notebook para a pasta .data
CSV_PATH = os.path.join('..', 'data', 'cloud_input', 'water_potability_without_target.csv')
```

```python
# Leitura do csv
df = pd.read_csv(CSV_PATH)
```

```python
df.head(30)
```

```python
# Nome da coleção
collection_name = 'devops-eml1-ex2'
```

```python
# Upload linha a linha
for i, row in df.iterrows():
    doc_ref = db.collection(collection_name).document(str(i))  # ID opcional
    doc_ref.set(row.to_dict())

print("✅ Upload concluído com sucesso para a coleção:", collection_name)
```
